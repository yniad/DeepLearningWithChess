```python
import numpy as np
import chess
import chess.pgn
import io
import pickle
```

```python
mapeo={
        '.':np.array([0,0,0,0,0],dtype=np.uint8),
        'r':np.array([0,1,0,0,1],dtype=np.uint8),
        'n':np.array([0,1,0,1,0],dtype=np.uint8),
        'b':np.array([0,1,0,1,1],dtype=np.uint8),
        'q':np.array([0,1,1,0,0],dtype=np.uint8),
        'k':np.array([0,1,1,0,1],dtype=np.uint8),
        'p':np.array([0,1,1,1,0],dtype=np.uint8),
        'R':np.array([1,0,0,0,1],dtype=np.uint8),
        'N':np.array([1,0,0,1,0],dtype=np.uint8),
        'B':np.array([1,0,0,1,1],dtype=np.uint8),
        'Q':np.array([1,0,1,0,0],dtype=np.uint8),
        'K':np.array([1,0,1,0,1],dtype=np.uint8),
        'P':np.array([1,0,1,1,0],dtype=np.uint8)
    }
```

```python
def codificar(board):
    b=str(board).replace(' ','').split('\n')
    a=np.zeros([8,8,5],dtype=np.uint8)
    
    for i,row in enumerate(b):
        for j,val in enumerate(row):
            if val=='.':
                continue
            a[i,j,:]=mapeo[val]
    return a
```

```python
def move_forward(board,moves,n):
    i=0
    while i<n:
        board.push(moves.pop(0)) #esta linea realiza el movimiento
        i+=1
```

```python
def get_random_values(N,top):
    r = np.arange(10,N)
    np.random.shuffle(r)
    r=r[:top]
    r.sort()
    for i in range(len(r)-1,0,-1):
        r[i]-=r[i-1]
    return r
```

```python
def save_file(data,save_path,player,winner,index):
    name=save_path+player+'_'+winner+'_'+str(index)+'.pkl'
    with open(name, 'wb') as outfile:
        pickle.dump(data, outfile, pickle.HIGHEST_PROTOCOL)
    print(name)
```

```python
def convertir():
    source='CCRL-4040.pgn'
    save_path='pgn/'
    #source='365chess_games.pgn'
    logfile = open(source, "r")
    i_b_b=0
    i_w_w=0
    i_b_w=0
    i_w_b=0
    file_b_b=0
    file_w_w=0
    file_b_w=0
    file_w_b=0
    
    max_v=10000
    
    control=0
    top=10

    data_w_w=np.zeros([10000,8,8,5],dtype=np.uint8) # turno w, gana w
    data_b_b=np.zeros([10000,8,8,5],dtype=np.uint8) # turno b, gana b
    data_w_b=np.zeros([10000,8,8,5],dtype=np.uint8) #turno w, gana b
    data_b_w=np.zeros([10000,8,8,5],dtype=np.uint8) #turno b, gana w
    
    pgn = open("CCRL-4040.pgn")
    game = chess.pgn.read_game(pgn)
    
    while game: 
        result = game.headers['Result']
        if '1-0' in result or '0-1' in result:
            N=len(list(game.mainline_moves()))
            values=get_random_values(N,top)
            board = game.board()
            moves=list(game.mainline_moves())
            if '1-0' in result:
                for v in values:
                    move_forward(board,moves,v)
                    if board.turn: #turno w
                        data_w_w[i_w_w,:]=codificar(board)
                        i_w_w+=1
                        if i_w_w>=max_v:
                            save_file(data_w_w,save_path,'w','w',file_w_w)
                            file_w_w+=1
                            i_w_w=0
                    else:
                        data_b_w[i_b_w,:]=codificar(board)
                        i_b_w+=1
                        if i_b_w>=max_v:
                            save_file(data_b_w,save_path,'b','w',file_b_w)
                            file_b_w+=1
                            i_b_w=0
                
            elif '0-1' in result:
                for v in values:
                    move_forward(board,moves,v)
                    if board.turn: #turno w
                        data_w_b[i_w_b,:]=codificar(board)
                        i_w_b+=1
                        if i_w_b>=max_v:
                            save_file(data_w_b,save_path,'w','b',file_w_b)
                            file_w_b+=1
                            i_w_b=0
                    else:
                        data_b_b[i_b_b,:]=codificar(board)
                        i_b_b+=1
                        if i_b_b>=max_v:
                            save_file(data_b_b,save_path,'b','b',file_b_b)
                            file_b_b+=1
                            i_b_b=0
        game = chess.pgn.read_game(pgn)
   

    if i_w_w>0:
        save_file(data_w_w[:i_w_w,:],save_path,'w','w',file_w_w)
        file_w_w+=1
        i_w_w=0
        data_w_w=[]

    elif i_b_b>0:
        save_file(data_b_b[:i_b_b,:],save_path,'b','b',file_b_b)
        file_b_b+=1
        i_b_b=0
        data_b_b=[]
    elif i_w_b>0:
        save_file(data_w_b[:i_w_b,:],save_path,'w','b',file_w_b)
        file_w_b+=1
        i_w_b=0
        data_w_b=[]
    elif i_b_w>0:
        save_file(data_b_w[:i_b_w,:],save_path,'b','w',file_b_w)
        file_b_w+=1
        i_b_w=0
        data_b_w=[]
```
